```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.countplot(x='total_bill', data=tips)
```

```python
sns.pointplot(x='total_bill',y='day',data=tips)
```

```python
sns.factorplot(x='smoker',y='tip', data=tips,kind='box')
```

```python
sns.swarmplot(x='smoker',y='tip',data=tips,hue='sex',palette='cool')
```

```python
sns.stripplot(x='smoker',y='tip',data=tips,hue='sex',palette='cool')
```

```python
sns.lineplot(x='smoker',y='tip',data=tips,hue='sex',palette='cool')
```

```python
sns.violinplot(x='smoker',y='tip',data=tips,hue='sex',palette='cool')
```
